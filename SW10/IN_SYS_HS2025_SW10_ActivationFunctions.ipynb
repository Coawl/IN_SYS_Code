```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Defining common activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

# Generating input data
x = np.linspace(-10, 10, 1000)

# Applying activation functions
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_relu = relu(x)
y_leaky_relu = leaky_relu(x)

# Creating a Seaborn plot for each activation function
sns.set(style="whitegrid")
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('IN-SYS - Common Activation Functions in Neural Networks')

# Plot for Sigmoid
sns.lineplot(x=x, y=y_sigmoid, ax=axes[0, 0], color='blue')
axes[0, 0].set_title('Sigmoid Function')
axes[0, 0].set_xlabel('Input')
axes[0, 0].set_ylabel('Output')

# Plot for Tanh
sns.lineplot(x=x, y=y_tanh, ax=axes[0, 1], color='green')
axes[0, 1].set_title('Tanh Function')
axes[0, 1].set_xlabel('Input')
axes[0, 1].set_ylabel('Output')

# Plot for ReLU
sns.lineplot(x=x, y=y_relu, ax=axes[1, 0], color='red')
axes[1, 0].set_title('ReLU Function')
axes[1, 0].set_xlabel('Input')
axes[1, 0].set_ylabel('Output')

# Plot for Leaky ReLU
sns.lineplot(x=x, y=y_leaky_relu, ax=axes[1, 1], color='purple')
axes[1, 1].set_title('Leaky ReLU Function')
axes[1, 1].set_xlabel('Input')
axes[1, 1].set_ylabel('Output')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
