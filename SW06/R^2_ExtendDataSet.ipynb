# Extend CaliforniaHousingCut Data Set

```python
import pandas as pd
import numpy as np

# Datei einlesen
df = pd.read_csv("CaliforniaHousingCut.csv")

# Zusätzliche Spalten erzeugen
np.random.seed(42)
df["Postal_Code"] = np.random.randint(90000, 96999, len(df))   # realistische CA-Postleitzahlen
df["Extra_Noise"] = np.random.normal(0, 1, len(df))            # normales Rauschen
df["Random"] = np.random.uniform(0, 100, len(df))              # gleichverteilte Zufallswerte

# Neue Spalte: Average_NotBedrooms = Average Rooms - Average Bedrooms
df["Average_NotBedrooms"] = df["Average Rooms"] - df["Average Bedrooms"]

# Neue Datei speichern
df.to_csv("CaliforniaHousingCut_extended.csv", index=False)

print("✅ Datei 'CaliforniaHousingCut_extended.csv' erfolgreich erstellt!")
```
