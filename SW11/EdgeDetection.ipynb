```python
import cv2
import matplotlib.pyplot as plt

# put your image in the folder in .png or .jpg format and load it with imread function
image = cv2.imread("SLAZ_2021.png", cv2.IMREAD_GRAYSCALE)

# Step 1: Denoising with Gaussian Blur, recommended for opreprocessing before edge detection
image_blur = cv2.GaussianBlur(image, (5, 5), 1)

# Step 2: Contrast Enhancement
image_equalized = cv2.equalizeHist(image_blur)

# Step 3: Edge Detection
edges = cv2.Canny(image_equalized, 100, 200)

# Visualization
plt.figure(figsize=(12, 8))

# Original Image
plt.subplot(1, 4, 1)
plt.imshow(image, cmap="gray")
plt.title("Original Image")
plt.axis("off")

# Preprocessed Image
plt.subplot(1, 4, 2)
plt.imshow(image_blur, cmap="gray")
plt.title("Blurred Image")
plt.axis("off")

# Preprocessed Image
plt.subplot(1, 4, 3)
plt.imshow(image_equalized, cmap="gray")
plt.title("High-contrast Image")
plt.axis("off")

# Edge Detection Result
plt.subplot(1, 4, 4)
plt.imshow(edges, cmap="gray")
plt.title("Edges")
plt.axis("off")

plt.tight_layout()
plt.savefig("SLAZ_edges.svg", format="svg")
plt.show()
```
