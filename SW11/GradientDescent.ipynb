```python
import numpy as np
import matplotlib.pyplot as plt

# Define the loss function and its gradient
def f(x):
    return x**2

def f_prime(x):
    return 2 * x
```

```python
# Sample point
x_sample = 5
y_sample = f(x_sample)
slope = f_prime(x_sample)

# Tangent line at x = 5
x_tangent = np.linspace(x_sample - 2, x_sample + 2, 100)
y_tangent = slope * (x_tangent - x_sample) + y_sample

# Plot the function and the tangent
x_plot = np.linspace(-6, 6, 500)
y_plot = f(x_plot)

plt.figure(figsize=(8, 6))
plt.plot(x_plot, y_plot, label="$f(x) = x^2$", color='blue')

plt.scatter(x_sample, y_sample, color='red', label=f"Sample Point ($x={x_sample}$, $f(x)={y_sample}$)", zorder=5)
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.title("Loss Function $f(x) = x^2$ with sample point at $x=5$")
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
plt.legend()
plt.grid(True)
plt.savefig("GradientExample_01.svg", format="svg")
plt.show()
```

```python
# Sample point
x_sample = 5
y_sample = f(x_sample)
slope = f_prime(x_sample)

# Tangent line at x = 5
x_tangent = np.linspace(x_sample - 2, x_sample + 2, 100)
y_tangent = slope * (x_tangent - x_sample) + y_sample

# Plot the function and the tangent
x_plot = np.linspace(-6, 6, 500)
y_plot = f(x_plot)

plt.figure(figsize=(8, 6))
plt.plot(x_plot, y_plot, label="$f(x) = x^2$", color='blue')
plt.plot(x_tangent, y_tangent, label="Tangent at $x=5$", color='orange', linestyle='--')
plt.scatter(x_sample, y_sample, color='red', label=f"Sample Point ($x={x_sample}$, $f(x)={y_sample}$)", zorder=5)
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.title("Loss Function $f(x) = x^2$ and its Tangent at $x=5$ with Gradient (slope) = 10")
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
plt.legend()
plt.grid(True)
plt.savefig("GradientExample_02.svg", format="svg")
plt.show()
```

```python
# Gradient Descent parameters
learning_rate = 0.1
initial_x = 5  # Starting point
iterations = 20

# Lists to store values for plotting
x_values = [initial_x]
f_values = [f(initial_x)]

# Perform Gradient Descent
x = initial_x
for i in range(iterations):
    gradient = f_prime(x)
    x = x - learning_rate * gradient  # Update x
    x_values.append(x)
    f_values.append(f(x))

# Plotting the function and Gradient Descent steps
x_plot = np.linspace(-6, 6, 500)
y_plot = f(x_plot)

fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Plot the function
ax[0].plot(x_plot, y_plot, label="$f(x) = x^2$")
ax[0].scatter(x_values, f_values, color='red', label='Steps', zorder=5)
for i, (x_step, y_step) in enumerate(zip(x_values, f_values)):
    ax[0].text(x_step, y_step + 0.5, f"{i}", fontsize=9)
ax[0].set_title("Gradient Descent on $f(x) = x^2$")
ax[0].set_xlabel("$x$")
ax[0].set_ylabel("$f(x)$")
ax[0].grid(True)
ax[0].legend()

# Plot the convergence of f(x) values
ax[1].plot(range(len(f_values)), f_values, marker='o', label='$f(x)$ values')
ax[1].set_title("Convergence of $f(x)$")
ax[1].set_xlabel("Iteration")
ax[1].set_ylabel("$f(x)$")
ax[1].legend()
ax[1].grid(True)
plt.tight_layout()
plt.savefig("GradientExample_03.svg", format="svg")
plt.show()
```
