```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
```

```python
# Load the dataset
data = pd.read_csv('three_colors_dataset.csv')
```

```python
# Display the first few rows of the dataset to understand its structure
data.head()
```

```python
# Convert this dataframe into numpy arrays

color_map = {'Red': 0, 'Blue': 1, 'Green': 2}
y = data['Color'].map(color_map).values 

# map function will apply mapping dictionary
# .values method returns values in the form of numpy array

# Features
X = data[['x', 'y']].values

print("X:\n", X)
print("y:", y)
```

```python
clf = DecisionTreeClassifier(
    criterion = 'gini',    # Use Gini impurity for splits
    random_state = 42,     # For reproducible results
    max_depth = 3          # Limit tree depth for visualization
)
```

```python
clf.fit(X, y)
```

```python
feature_names = ['X1', 'X2']
class_names = ['red', 'blue', 'green']

plt.figure(figsize = (5, 7))
plot_tree(clf, 
          feature_names=feature_names,
          class_names=class_names,
          filled=False,         # No colors in the tree leaves
          rounded=False,        # Squared node corners
          proportion=False)     # Show counts
plt.title("Decision Tree for Three Colors Dataset")
plt.show()
```
