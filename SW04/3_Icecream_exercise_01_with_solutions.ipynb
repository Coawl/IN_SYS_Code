### Task 1: Add Packaging Sizes and Calorie Information

Add three packaging sizes for each flavor:  
- mini (50g) - % of base price  *(decide what percentage you will give here)*
- standard (100g) - base price (already given)  
- maxi (200g) - % of base price   *(decide what percentage you will give here)*

Add calories per package using these base calories per **50g**:

- Vanilla: 120 calories
- Chocolate: 140 calories
- Coffee: 110 calories
- Lemon: calories  *(decide yourself on how many calories)*
- Strawberry: calories   *(decide yourself on how many calories)*

You are supposed to use commands to **add into the existing, given dictionary**, not to create a new dictionary.  
In the end, you will have three packaging sizes per flavor 

```python
# CALORIE INFORMATION
calorie_base = {
    'vanilla': 120,
    'chocolate': 140,
    'coffee': 110,
    'lemon': 90,
    'strawberry': 115
}

# ADD PACKAGING SIZES AND CALORIES
for flavor in ice_cream_dict:
    base_price = ice_cream_dict[flavor]['base_price_chf']
    base_calories = calorie_base[flavor]
    
    ice_cream_dict[flavor]['packaging'] = {
        'mini': {
            'weight_g': 50,
            'price_chf': round(base_price * 0.6, 1),
            'calories': base_calories
        },
        'standard': {
            'weight_g': 100,
            'price_chf': base_price,
            'calories': base_calories * 2
        },
        'maxi': {
            'weight_g': 200,
            'price_chf': round(base_price * 1.6, 1),
            'calories': base_calories * 4
        }
    }

print("âœ… Packaging and calories added to dictionary")
```

### Task 2: Visualize the dictionary structure in a tree
Also, count how many nested dictionaries are there and how many levels

```python
def print_dict_tree(d, indent=0):
    """Print dictionary structure as a tree"""
    for key, value in d.items():
        print('  ' * indent + 'â”œâ”€â”€ ' + str(key))
        if isinstance(value, dict):
            print_dict_tree(value, indent + 1)

print("\nğŸŒ³ Dictionary Structure Tree:")
print_dict_tree(ice_cream_dict)
```

```python
import pandas as pd

pd.set_option('display.width', 1000)  # Increase total width
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.expand_frame_repr', False)  # Don't wrap to multiple lines

# CONVERT TO DATAFRAME
records = []
for flavor, details in ice_cream_dict.items():
    for size, package_info in details['packaging'].items():
        records.append({
            'flavor': flavor,
            'package_size': size,
            'weight_g': package_info['weight_g'],
            'price_chf': package_info['price_chf'],
            'calories': package_info['calories'],
            'base_price_chf': details['base_price_chf'],
            'description': details['description']
        })

df = pd.DataFrame(records)

print("\nâœ… Converted to DataFrame:")
print(f"DataFrame shape: {df.shape}")
print(df.head())
```

### Task 4: Save this into CSV file
